```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("Position_Salaries.csv")
X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
```

```python
plt.scatter(df['Level'],df['Salary'], color= 'blue')
plt.show()
```

```python
lreg = LinearRegression()
lreg.fit(X,y)
```

```python
y_pred = lreg.predict(X)
```

```python
r2_score(y, y_pred)*100
```

```python
plt.scatter(X,y, color= 'blue')
plt.plot(X, y_pred , color= 'red')
plt.title("Linear Regression Level vs Salary")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
lreg.predict([[6.5]])
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X)
preg1 = LinearRegression()
preg1.fit(X_poly,y)
```

```python
X_poly
```

```python
preg1.predict(poly_reg.fit_transform([[6.5]]))
```

```python
yp_pred = preg1.predict(X_poly)
plt.scatter(X,y, color= 'blue')
plt.plot(X, yp_pred , color= 'red')
plt.title("Polynominal Regression Level vs Salary")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
r2_score(y, yp_pred)*100
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
preg2 = LinearRegression()
preg2.fit(X_poly,y)
```

```python
yp_pred = preg2.predict(X_poly)
plt.scatter(X,y, color= 'blue')
plt.plot(X, yp_pred , color= 'red')
plt.title("Polynominal Regression Level vs Salary")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
r2_score(y, yp_pred)*100
```

```python
preg2.predict(poly_reg.fit_transform([[6.5]]))
```
